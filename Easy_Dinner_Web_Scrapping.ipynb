### Collecting Data From EasyDinner Website

#### Restaurants in Hyderabad

```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import time

Restaurant_name = [] 
Restaurant_style = []
Price = []
Discount = []
Address = []
pagenum=[]  


total_time = time.time()

for i in range(1, 101): 
    start_time = time.time()
    URL = f"https://www.eazydiner.com/restaurants?city=hyderabad&location=hyderabad&page={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.text, "html.parser")

    for x in soup.find_all('div', attrs={'class':'relative pointer listing_touch_box__00o7f'}):
        Rest_name = x.find('div', attrs={'class':'flex align-v-center flex-between'})
        Rest_style = x.find('div', attrs={'class':'flex align-v-center ellipsis listing_multi_cuisines__zLLXK'}) 
        price = x.find('div', attrs={'class':'flex align-v-center'})
        discount = x.find('div', attrs={'class':'ellipsis listing_offer_text__LXKw1'})
        address = x.find('div', attrs={'class':'ellipsis'})


        Restaurant_name.append(Rest_name.text if Rest_name else np.nan)
        Address.append(address.text if address else np.nan)
        Restaurant_style.append(Rest_style.text if Rest_style else np.nan)
        Price.append(price.text if price else np.nan)
        Discount.append(discount.text if discount else np.nan)
        
        pagenum.append(i)

    print(f'Page {i} completed in {time.time()-start_time:.2f} seconds')

print("Total Time Completed in seconds", str(time.time()-total_time))
```

```python
df = pd.DataFrame({"Restaurant_Name":Restaurant_name,"Restaurant_Style":Restaurant_style,"Price":Price,"Discount":Discount,"Address":Address})
```

```python
df['Rating'] = df['Restaurant_Name'].str.extract(r'(\d\.\d)$')
```

```python
df.head()
```

```python
df['Restaurant_Name'] = df['Restaurant_Name'].apply(lambda x : re.sub(r'\d+\.\d+','',x)) 
```

```python
df.head()
```

```python
df['Address'] = df['Address'].str.replace(r'\|.*', '', regex=True).str.strip()
```

```python
df['Locality'] = df['Address'].str.replace(r'\s+\S+$', '', regex=True).str.rstrip(', ')
```

```python
df['City'] = df['Address'].str.extract(r'(\w+)$')
```

```python
df.head()
```

```python
df.drop(columns=['Address'], inplace=True, errors='ignore')
```

```python
df.head()
```

```python
df['Price'] = df['Price'].str.replace("₹", "", regex=False).str.strip()
```

```python
df.head()
```

```python
df['Price_Per_Two'] = df['Price'].str.extract(r'(\d+)').astype('int')
```

```python
df
```

```python
df
```

```python
df.drop(columns=['Price'], inplace=True, errors='ignore')
```

```python
df
```

```python
df['Price_For_One'] = (df['Price_Per_Two']/2).astype(int)
```

```python
df
```

```python
df = df[['Restaurant_Name','Restaurant_Style','Price_Per_Two','Price_For_One','Discount','Rating','Locality','City']]
```
